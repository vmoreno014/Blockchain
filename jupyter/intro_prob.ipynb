# Probabilidad

### Probabilidad de que ocurra un evento
**p** (evento) = nº veces (evento ocurre) / nº veces probamos

### Ley de los numeros grandes
Tras un número grande de pruebas, nos podemos fiar del resultado  

### Ejemplos
- *Una moneda:*  
p(cara) = 0.5 y p(cruz) = 0.5  

- *Un dado:*  
p(1 or 2 or 3 or 4 or 5 or 6) = 1/6 = 0.166

### Evento dependiente o independiente
Supongamos que tengo una bolsa con 5 canicas de las cuales 3 son rojas y 2 azules. Vamos a sacar una canica cualquiera:
p(roja) = 3/5 = 0.6
p(azul) = 2/5 = 0.4

1. Saco una canica roja, pero no la vuelvo a meter.
Las probabilidades para la siguiente vez que saque canicas van  a cambiar.
p(roja) = 2/4 = 0.5
p(azul) = 2/4 = 0.5
Por tanto, la probabilidad del segundo intento **depende** del resultado del primer intento. **Evento dependiente**

2. Saco una canica roja, pero la vuelvo a meter.
Mismas probabilidades.
p(roja) = 3/5 = 0.6
p(azul) = 2/5 = 0.4
Por tanto, la probabilidad del segundo intento **no depende** del resultado del primer intento. **Evento independiente**

### Mutuamente exclusivo
1. Dos eventos, **no son mutuamente exclusivos** cuando ambos pueden ocurrir a la vez.  
(*Moneda*) Sacar cara o sacar cruz.   
(*Dado*) Sacar un número par y mayor que dos.

2. Dos eventos, son **mutuamente exclusivos** cuando ambos no pueden ocurrir a la vez.  
(*Moneda*) Sacar cruz y cara en el mismo lanzamiento.  
(*Dado*) Sacar un número par menor que dos.


### Fórmulas
P(A or B) = P(A) + P(B) - P(A and B), *eventos no mutuamente exclusivos*  
P(A or B) = P(A) + P(B), *eventos mutuamente exclusivos*

```python
def prob_a_or_b(a, b, all_possible_outcomes):
    prob_a = len(a)/len(all_possible_outcomes)
    prob_b = len(b)/len(all_possible_outcomes)
    inter = a.intersection(b)
    prob_inter = len(inter)/len(all_possible_outcomes)
    return (prob_a + prob_b - prob_inter)

# rolling a die once and getting an even number or an odd number
evens = {2, 4, 6}
odds = {1, 3, 5}
all_possible_rolls = {1, 2, 3, 4, 5, 6}
res = prob_a_or_b(evens, odds, all_possible_rolls)
print("{:.2f}% obtener un numero par o impar".format(res*100)) # 100%

# rolling a die once and getting an odd number or a number greater than 2
odds = {1, 3, 5}
greater_than_two = {3, 4, 5, 6}
all_possible_rolls = {1, 2, 3, 4, 5, 6}
res = prob_a_or_b(odds, greater_than_two, all_possible_rolls)
print("{:.2f}% obtener un numero impar o mayor que 2, en el primer intento".format(res*100)) # 83%

# selecting a diamond card or a face card from a standard deck of cards
diamond_cards = {'ace_diamond', '2_diamond', '3_diamond', '4_diamond', '5_diamond', '6_diamond', '7_diamond', '8_diamond', '9_diamond', '10_diamond', 'jack_diamond', 'queen_diamond', 'king_diamond'}
face_cards = {'jack_diamond', 'jack_spade', 'jack_heart', 'jack_club', 'queen_diamond', 'queen_spade', 'queen_heart', 'queen_club', 'king_diamond', 'king_spade', 'king_heart', 'king_club'}
# all cards in a deck representing the entire sample space
all_possible_cards = {'ace_diamond', '2_diamond', '3_diamond', '4_diamond', '5_diamond', '6_diamond', '7_diamond', '8_diamond', '9_diamond', '10_diamond', 'jack_diamond', 'queen_diamond', 'king_diamond', 'ace_heart', '2_heart', '3_heart', '4_heart', '5_heart', '6_heart', '7_heart', '8_heart', '9_heart', '10_heart', 'jack_heart', 'queen_heart', 'king_heart', 'ace_spade', '2_spade', '3_spade', '4_spade', '5_spade', '6_spade', '7_spade', '8_spade', '9_spade', '10_spade', 'jack_spade', 'queen_spade', 'king_spade', 'ace_club', '2_club', '3_club', '4_club', '5_club', '6_club', '7_club', '8_club', '9_club', '10_club', 'jack_club', 'queen_club', 'king_club'}
res = prob_a_or_b(diamond_cards, face_cards, all_possible_cards)
print("{:.2f}% obtener una carta de diamantes o una carta de sota, caballo o rey".format(res*100)) # 42%
```

### Probabilidad condicionada
Sacar una canica roja sabiendo que la primera a sido azul  
p(roja segunda | azul primera) = p(roja segunda)
p(azul primera | roja segunda) = p(azul primera)

### Regla de la multiplicacion
P(A and B) = P(A) * P(B | A)


### Teorema de Bayes
P(B|A) = ( P(A|B) * P(B) ) / P(A)

### Ejercicio  


ST = Step throat  
P(ST) = **0.20**
P(+|ST) = **0.85**, (*sabiendo que tiene ST, que de positivo*)
P(+|NO ST) = **0.02**, (*sabiendo que NO tiene ST, que de positivo*)

#### 1. Calcular P(NO ST)  
P(NO ST) = 1 - P(ST) = 1 - 0.20 = **0.80**

#### 2. Calcular P(-|ST)  
*Sabiendo que tiene ST, que de negativo*  
P(-|ST) = 1 - P(+|ST) = 1 - 0.85 = **0.15**  

#### 3. Calcular P(-|NO ST)  
*Sabiendo que NO tiene ST, que de negativo*  
P(-|NO ST) = 1 - P(+|NO ST) = 1 - 0.02 = **0.98**  

#### 4. Calcular P(ST and +)  
*Que tenga ST y sea positivo*  
P(ST and +) = P(ST) * P(+|ST) = 0.20 * 0.85 = **0.17**

#### 5. Calcular P(ST and -)  
*Que tenga ST y sea negativo*  
P(ST and -) = P(ST) * P(-|ST) = 0.20 * 0.15 = **0.03**

#### 6. Calcular P(NO ST and +)  
*Que NO tenga ST y sea positivo*  
P(NO ST and +) = P(NO ST) * P(+|NO ST) = 0.80 * 0.02 = **0.016**

#### 7. Calcular P(NO ST and -)  
*Que NO tenga ST y sea negativo*  
P(NO ST and -) = P(NO ST) * P(-|NO ST) = 0.80 * 0.98 = **0.784**

#### 8. Calcular P(+)  
*Ser positivo*  
P(+) = P(ST and +) + P(NO ST and+) = 0.17 + 0.016 = **0.186**

#### 9. Calcular P(-)  
*Ser negativo*  
P(-) = P(ST and -) + P(NO ST and -) = 0.03 + 0.784 = **0.814**  
P(-) = 1 - P(+) = 1 - 0.186 = 0.814

#### 10. Calcular P(ST|+)  
*Sabiendo que eres positivo, que tengas ST*  
P(ST|+) = (P(+|ST) * P(ST)) / P(+) = (0.85 * 0.20) / 0.186 = **0.914**

#### 11. Calcular P(ST|-)  
*Sabiendo que eres negativo, que tengas ST*  
P(ST|-) = (P(-|ST) * P(ST)) / P(-) = (0.15 * 0.20) / 0.814 = **0.036**

#### 12. Calcular P(NO ST|+)  
*Sabiendo que es positivo, que NO tengas ST*  
P(NO ST|+) = (P(+|NO ST) * P(NO ST)) / P(+) = (0.02 * 0.80) / 0.186 = **0.08**

#### 13. Calcular P(NO ST|-)  
*Sabiendo que es positivo, que NO tengas ST*  
P(NO ST|-) = (P(-|NO ST) * P(NO ST)) / P(-) = (0.98 * 0.80) / 0.814 = **0.96**
